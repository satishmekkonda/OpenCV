## Child Image

```python
child = cv2.imread('DATA/Portion.jpg')
child = cv2.cvtColor(child, cv2.COLOR_BGR2RGB)
print(child.shape)
plt.imshow(child)
```

```python
##Note below are strings, we will later use eval and cobnvert to functions
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR', 'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
my_method = eval('cv2.TM_SQDIFF')
res = cv2.matchTemplate(first, child, my_method)
plt.imshow(res)
```

## Detecting Child-Image with Parent Image

```python
for m in methods:
    
    first_copy = first.copy()
    
    method = eval(m)
    
    ##Template Matching
    res = cv2.matchTemplate(first_copy, child, method)                   
    
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc   ##(x,y)
        
    else:
        top_left = max_loc
        
    height, width, channels = child.shape
    
    bottom_right = (top_left[0]+width, top_left[1]+height)
    
    cv2.rectangle(first_copy, top_left, bottom_right, (255,0,0), 10)
    
    
    #Plotting Images
    plt.subplot(121)
    plt.imshow(res)
    plt.title('Heat Map Image')
    plt.subplot(122)
    plt.imshow(first_copy)
    plt.title('Detected Image')
    
    plt.suptitle(m)
    plt.show()
    
    print('\n')
    print('\n')   
```

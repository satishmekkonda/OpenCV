```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
```

## Draw Shapes

```python
## Rectangles
cv2.rectangle(blank_img, pt1=(384, 10), pt2=(500,150), color=(0,255,0), thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
##Square
cv2.rectangle(blank_img, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness=10)
plt.imshow(blank_img)
```

```python
## Circles
cv2.circle(blank_img, center=(100, 100), radius=50, color=(255, 0, 0), thickness = 8)
plt.imshow(blank_img)
```

```python
## Filling Shapes

cv2.rectangle(blank_img, pt1=(100, 480), pt2=(200,400), color=(0,255,0), thickness=-1)
cv2.rectangle(blank_img, pt1=(400,200), pt2=(480,300), color=(0,0,255), thickness=-1)
cv2.circle(blank_img, center=(400, 400), radius=50, color=(255, 0, 0), thickness = -1)
plt.imshow(blank_img)
```

## Drawing Lines

```python
cv2.line(blank_img, pt1=(0,0), pt2=(400,400), color=(102, 255, 255), thickness=5)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1=(512,512), pt2=(0,500), color=(102, 255, 255), thickness=5)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1=(512,0), pt2=(0,500), color=(102, 255, 255), thickness=5)
plt.imshow(blank_img)
```

## Drawing Fonts

```python
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(blank_img, text = 'Hello', org=(10, 500),
            fontFace = font,
            fontScale = 4,
            color = (255,255,255),
            thickness = 3,
            lineType = cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(blank_img, text = 'Hello', org=(100, 100),
            fontFace = font,
            fontScale = 2,
            color = (255,255,255),
            thickness = 2,
            lineType = cv2.LINE_AA)
plt.imshow(blank_img)
```

## Drawing Polygons

```python
new_img = np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(new_img)
```

```python
vertices = np.array([[100,300], [200,200], [400,300], [200,400]], dtype=np.int32)
print(vertices.shape)
vertices
```

```python
pts = vertices.reshape(-1,1,2)
print(pts.shape)
pts
```

```python
pts = vertices.reshape((-1,1,2))
print(pts.shape)
pts
```

```python
## Reshaping is done because cv2 has 3 channels
```

```python
cv2.polylines(blank_img, [pts], isClosed = True, color = (255,0,0), thickness=5)
plt.imshow(blank_img)
```

```python
cv2.polylines(new_img, [pts], isClosed = True, color = (255,0,0), thickness=5)
plt.imshow(new_img)
```

```python
new_img2 = np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(new_img2)

cv2.polylines(new_img2, [vertices], isClosed = True, color = (255,0,0), thickness=5)
plt.imshow(new_img2)
```

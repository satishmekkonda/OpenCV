```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
full = cv2.imread('DATA/sammy.jpg')
full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)
plt.imshow(full)
```

```python
face = cv2.imread('DATA/sammy_face.jpg')
face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
plt.imshow(face)
```

```python
print(full.shape)
print(face.shape)
```

```python
##Note below are strings, we will later use eval and cobnvert to functions
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR', 'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
my_method = eval('cv2.TM_CCOEFF')
res = cv2.matchTemplate(full, face, my_method)
plt.imshow(res)
```

```python
my_method = eval('cv2.TM_CCOEFF_NORMED')
res = cv2.matchTemplate(full, face, my_method)
plt.imshow(res)
```

```python
my_method = eval('cv2.TM_CCORR')
res = cv2.matchTemplate(full, face, my_method)
plt.imshow(res)
```

```python
my_method = eval('cv2.TM_CCORR_NORMED')
res = cv2.matchTemplate(full, face, my_method)
plt.imshow(res)
```

```python
my_method = eval('cv2.TM_SQDIFF')
res = cv2.matchTemplate(full, face, my_method)
plt.imshow(res)
```

```python
my_method = eval('cv2.TM_SQDIFF_NORMED')
res = cv2.matchTemplate(full, face, my_method)
plt.imshow(res)
```

```python
for m in methods:
    
    full_copy = full.copy()
    
    method = eval(m)
    
    ##Template Matching
    res = cv2.matchTemplate(full_copy, face, method)                   
    
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc   ##(x,y)
        
    else:
        top_left = max_loc
        
    height, width, channels = face.shape
    
    bottom_right = (top_left[0]+width, top_left[1]+height)
    
    cv2.rectangle(full_copy, top_left, bottom_right, (255,255,255), 10)
    
    
    #Plotting Images
    plt.subplot(121)
    plt.imshow(res)
    plt.title('Heat Map Image')
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title('Detected Image')
    
    plt.suptitle(m)
    plt.show()
    
    print('\n')
    print('\n')   
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img, cmap = None):
    fig = plt.figure(figsize = (10, 8))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
img = cv2.imread('DATA/giraffes.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

display_img(img)
```

## Binary Threshold on a Image

```python
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
ret
```

```python
display_img(thresh1)
```

```python
img = cv2.imread('DATA/giraffes.jpg', 0)
ret, thresh2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
display_img(thresh2)
```

```python
img = cv2.imread('DATA/giraffes.jpg', 0)
ret, thresh3 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
display_img(thresh3, cmap='gray')
```

## Covert Coorspace to HSV

```python
img = cv2.imread('DATA/giraffes.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
display_img(img)
```

## Lowpass Filter with a 4x4 kernel, filled with values of 0.01 and then use 2D conv to blur image

```python
kernel = np.ones(shape=(4,4), dtype=np.float32)/10
kernel
```

```python
img = cv2.imread('DATA/giraffes.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

result = cv2.filter2D(img, -1, kernel)
display_img(result)
```

## Horizonat Sobel Filter

```python
img = cv2.imread('DATA/giraffes.jpg')

sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize = 5)

display_img(sobelx)
```

```python
img = cv2.imread('DATA/giraffes.jpg', 0)

sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize = 5)

display_img(sobelx, cmap='gray')
```

## COLOR HISTOGRAMS

```python
img = cv2.imread('DATA/giraffes.jpg')

color = ['b', 'g', 'r']

for i, col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histr)
    
plt.title('My Histo')
```

```python
img = cv2.imread('DATA/giraffes.jpg')

color = ['b', 'g', 'r']

for i, col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histr, color=col)
    
plt.title('My Histo')
```

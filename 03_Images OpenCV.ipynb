## Resize Image

```python
new_img = cv2.resize(fix_img, (1000, 400))
new_img.shape
```

```python
plt.imshow(new_img)
```

```python
w_ration = 0.5
h_ration = 0.5

new_img_2 = cv2.resize(fix_img, (0,0), fix_img, w_ration, h_ration)
print(new_img.shape)
plt.imshow(new_img_2)
```

```python
## 80% and 20%
w_ration = 0.8 
h_ration = 0.2

new_img_2 = cv2.resize(fix_img, (0,0), fix_img, w_ration, h_ration)
print(new_img.shape)
plt.imshow(new_img_2)
```

## Flipping Image

```python
## Flipping Image Vertically

flip_img = cv2.flip(fix_img, 0)
print(flip_img.shape)
plt.imshow(flip_img)
```

```python
## Flipping Image Horizantally

flip_img = cv2.flip(fix_img, 1)
print(flip_img.shape)
plt.imshow(flip_img)
```

```python
## Flipping Image Both

flip_img = cv2.flip(fix_img, -1)
print(flip_img.shape)
plt.imshow(flip_img)
```

```python
gray = np.float32(gray_flat_chess)
gray
```

## Corner Detection Algorithm

```python
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.004)

dst = cv2.dilate(dst, None)
```

```python
#Applying on original image
flat_chess[dst>0.01*dst.max()] = [255,0,0]  #Marking as Red
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chess)

dst = cv2.cornerHarris(src = gray, blockSize = 2, ksize = 3, k = 0.04)

dst = cv2.dilate(dst, None)

real_chess[dst>0.01*dst.max()] = [255, 0, 0]

plt.imshow(real_chess)
```

## PART - II

```python
real_chess = cv2.imread('DATA/real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)

flat_chess = cv2.imread('DATA/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)

gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2GRAY)
gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess, 5, 0.01, 10)  #Negative Numbers return ALl CORNERS
corners
```

```python
corners = np.int0(corners)
corners
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(flat_chess, (x, y), 3, (255, 0, 0), -1)
    
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess, 64, 0.01, 10)  #Negative Numbers return ALl CORNERS
corners = np.int0(corners)
for i in corners:
    x, y = i.ravel()
    cv2.circle(flat_chess, (x, y), 3, (255, 0, 0), -1)
    
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_chess, 80, 0.01, 10)  #Negative Numbers return ALl CORNERS
corners = np.int0(corners)
for i in corners:
    x, y = i.ravel()
    cv2.circle(real_chess, (x, y), 3, (255, 0, 0), -1)
    
plt.imshow(real_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_chess, 100, 0.01, 10)  #Negative Numbers return ALl CORNERS
corners = np.int0(corners)
for i in corners:
    x, y = i.ravel()
    cv2.circle(real_chess, (x, y), 3, (255, 0, 0), -1)
    
plt.imshow(real_chess)
```

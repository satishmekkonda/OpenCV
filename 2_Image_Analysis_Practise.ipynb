```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Color Photo Pixelled.jpg')
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

## FLIPPING

```python
new_img = img_rgb.copy()
plt.imshow(new_img)
```

```python
new_img = cv2.flip(new_img, 0)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(new_img, 1)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(new_img, 2)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(new_img, -1)
plt.imshow(new_img)
```

## Draw Rectangle

```python
plt.imshow(img_rgb)
```

```python
pt1 = (450,350)
pt2 = (750, 700)

cv2.rectangle(img_rgb, pt1 = pt1, pt2 = pt2, color=(255,0,0), thickness=10)
plt.imshow(img_rgb)
```

## Draw Traingle

```python
vertices = np.array([[1000, 250], [800, 700], [1200, 700]], dtype = np.int32)
print(vertices.shape)
vertices
```

```python
pts = vertices.reshape([-1,1,2])
print(pts.shape)
pts
```

```python
cv2.polylines(img_rgb, [pts], isClosed=True, color=(0,0,255), thickness=20)
plt.imshow(img_rgb)
```

## Fill Poly

```python
#Filling Traingle

vertices = np.array([[1000, 350], [850, 650], [1150, 650]], dtype = np.int32)
print(vertices.shape)

pts = vertices.reshape([-1,1,2])
print(pts.shape)


cv2.fillPoly(img_rgb, [pts], color=(0,0,255))
plt.imshow(img_rgb)
```
